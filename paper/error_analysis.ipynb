```python
import pandas as pd
import numpy as np
import seaborn as sns
import json
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 500)
```

# ASC Error Analysis

```python
task="ae"
domains=["rest", "laptop"]

preds = pd.DataFrame()
for domain in domains:
    p = pd.DataFrame(json.load(open(f"../run/rob_{task}/{domain}/1/predictions.json")))
    print(len(p))
    preds = preds.append(p).reset_index()

preds = preds.drop(["level_0","index"], axis=1)
preds["logits"] = preds.apply(lambda row: row.logits[:len(row.idx_map)], axis=1)
preds["label_idx"] = preds.logits.apply(lambda logit: [np.argmax(x) for x in logit])

print(len(preds))
preds.head()
```

```python
label_map = ['O','B','I']

def label_idx_mapping(row):
    _, raw_X, idx_map, label_idx = row
    # print(raw_X, idx_map, label_idx)
    labels = []
    for i in range(len(raw_X)):
        idx = idx_map.index(i)
        if idx >= len(label_idx):
            break
        label_id = label_idx[idx]
        label = label_map[label_id]
        labels.append(label)
    return labels

labelses = []
for _, row in preds.iterrows():
    # print(row)
    labelses.append(label_idx_mapping(row))
preds["pred_label"] = pd.Series(labelses)
print(len(preds))
preds
```

```python
inputs = pd.read_json(f"../ae/rest/test.json").T
inputs = inputs.append(pd.read_json(f"../ae/laptop/test.json").T).reset_index()
inputs
```

```python
df = preds.join(inputs)
df[df.label!=df.pred_label][["pred_label","label","sentence"]].sample(50).to_csv("ae_errors.csv")
```

# AE Error Analysis

```python
task="asc"
domain="rest"

preds = json.load(open(f"../run/rob_{task}/{domain}/1/predictions.json"))

data = []
for i in range(len(preds["logits"])):
    data.append({"logits": preds["logits"][i], "labels": preds['label_ids'][i]})

data = pd.DataFrame(data)
data
```

```python
def label_from_logits(logits):
    return logits.index(max(logits))

data["pred_label"]=data.logits.apply(label_from_logits)
data
```

```python
inputs = pd.read_json(f"../{task}/{domain}/test.json").T.reset_index().drop("index", axis=1)
label_lookup=['positive','negative','neutral']
inputs["labels"] = inputs.polarity.apply(lambda x: label_lookup.index(x))
inputs
```

```python
data = data.join(inputs, rsuffix="_y")
data[data.labels != data.labels_y]
```

```python
data[data.labels!=data.pred_label].sample(50).to_csv("asc_errors.csv")
```
